```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    filepath_or_buffer='/content/full_grouped.csv',
    header=0,
    names=['date','country','confirmed','deaths','recovered','active','newcases','newdeaths','newrecovered','region'],
    parse_dates=['date'])

monthDf = df.loc[(df['date'] >= '2020-06-01') & (df['date'] <= '2020-06-30')]
# monthDf.head()
```

```python
rncDf = monthDf.pivot_table(index=['date'],columns=['region'],values=['recovered'],aggfunc='sum')
# rncDf.head(20)
rncDf.plot()
```

```python
rncDf = monthDf.pivot_table(index=['date'],columns=['region'],values=['deaths'],aggfunc='sum')
# rncDf.head(20)
rncDf.plot()
```
